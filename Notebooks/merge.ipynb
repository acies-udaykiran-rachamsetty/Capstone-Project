```python
merged_df['date'] = pd.to_datetime(merged_df['date'])
merged_df = merged_df[merged_df['date'].dt.year != 2012]
```

```python
output = []

for column in merged_df.columns:

    null_values  = np.sum(pd.isna(merged_df[column]))
    unique = merged_df[column].nunique()
    column_data_type = merged_df[column].dtype

    output.append([column, null_values, unique, column_data_type])

output_df = pd.DataFrame(output, columns = ['column','null values', 'unique', 'datatype'])

describe_df = merged_df.describe().transpose() 
combine_df = pd.concat([ output_df.set_index('column'), describe_df], axis = 1, join='outer')

combine_df
```
