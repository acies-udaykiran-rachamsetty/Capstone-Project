```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
oil= pd.read_csv(r"..\Data\01_raw\oil.csv")
stores = pd.read_csv(r"..\Data\01_raw\stores.csv")
holidays = pd.read_csv(r"..\Data\01_raw\holidays_events.csv")
train = pd.read_csv(r"..\Data\01_raw\train.csv")
transactions = pd.read_csv(r"..\Data\01_raw\transactions.csv")
```

```python
# merged train and store using store number
merged_df = pd.merge(train, stores, on='store_nbr', how='outer')
```

```python
merged_df.sample(5)
```

```python
merged_df.shape
```

```python
output = []

for column in merged_df.columns:

    null_values  = np.sum(pd.isna(merged_df[column]))
    unique = merged_df[column].nunique()
    column_data_type = merged_df[column].dtype

    output.append([column, null_values, unique, column_data_type])

output_df = pd.DataFrame(output, columns = ['column','null values', 'unique', 'datatype'])

describe_df = merged_df.describe().transpose() 
combine_df = pd.concat([ output_df.set_index('column'), describe_df], axis = 1, join='outer')

combine_df
```

```python
# merged (train + store) and oil using date
merged_df = pd.merge(merged_df, oil, on='date', how='outer')
```

```python
merged_df.sample(5)
```

```python
merged_df['date'] = pd.to_datetime(merged_df['date'])
```

```python
np.sum(pd.isna(merged_df))
```

```python
merged_df['dcoilwtico'] = merged_df['dcoilwtico'].interpolate(method='linear')
```

```python
np.sum(pd.isna(merged_df))
```

```python
merged_df['dcoilwtico'][0] = 93.14
```

```python
merged_df.loc[merged_df['date'] == '2013-01-01', 'dcoilwtico'] = merged_df.loc[merged_df['date'] == '2013-01-01', 'dcoilwtico'].fillna(method='ffill')
```

```python
merged_df[merged_df['id'].isna() == True]
```

```python
# merged (train + store + oil ) and transaction using date + store number
transactions['date'] = pd.to_datetime(transactions['date'])
merged_df = pd.merge(merged_df, transactions, on=['date', 'store_nbr'], how='outer')
```

```python
np.sum(pd.isna(merged_df))
```

```python
holidays_subset = holidays[['date', 'type']]
```

```python
holidays_subset['date'] = pd.to_datetime(holidays_subset['date'])
merged_df = pd.merge(merged_df, holidays_subset, on='date', how='outer')
```

```python
np.sum(pd.isna(merged_df))
```

```python
merged_df = merged_df[merged_df['date'] <= '2017-08-15']
merged_df = merged_df[merged_df['date'].dt.year != 2012]
```

```python
blank = merged_df[merged_df['id'].isna() == True].index
blank.tolist()
merged_df = merged_df.drop(blank)
```

```python
merged_df['transactions'] = merged_df['transactions'].fillna(0)
```

```python
merged_df['type_y'] = merged_df['type_y'].fillna('No Holiday')
```

```python
merged_df['type_y'].unique()
```

```python
np.sum(pd.isna(merged_df))
```

```python
merged_df.rename(columns={'type_x': 'store_type'}, inplace=True)
merged_df.rename(columns={'type_y': 'holiday_type'}, inplace=True)
merged_df.rename(columns={'dcoilwtico': 'oil_price'}, inplace=True)
```

```python
merged_df.to_csv(r'..\Data\03_final\final.csv')
```
